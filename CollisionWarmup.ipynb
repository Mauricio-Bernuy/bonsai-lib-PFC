#Warm Up Disk

```python
#convert from text formattipsy.txt2tipsy('initial/collision_bulge.txt','initial/collision_bulge.tipsy')tipsy.txt2tipsy('initial/collision_disk.txt','initial/collision_disk.tipsy')tipsy.txt2tipsy('initial/collision_halo.txt','initial/collision_halo.tipsy')collision = tipsy.Stars('initial/collision_disk.tipsy')collision.append('initial/collision_bulge.tipsy')collision.append('initial/collision_halo.tipsy')print "Mass of system: %f" % sum(collision.mass)#reset timestampcollision.time = 0.0#save a new tipsy filecollision.save_tipsy('initial/cold_disk.tipsy')#save a figure of initial contidionscollision.save_figure('cold_disk',lim=3,pointsize=.05)#,nRed=31000)#display initial conditionsfrom IPython.core.display import ImageImage(filename='cold_disk.png')
```
